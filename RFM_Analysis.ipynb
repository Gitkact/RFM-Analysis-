## 1.2 Perform Univariate and Bivariate Analysis. Do exploratory data analysis. Check for Outlier

```python
def remove_outlier(col):
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
def univariateAnalysis_numeric(data,column):
    df_temp=data
    
    print("Description of " + column)
    print("----------------------------------------------------------------------------")
    print(df_temp[column].describe(),end=' ')
    print('')
    
    plt.figure()
    print("Distribution of " + column)
    print("----------------------------------------------------------------------------")
    sns.distplot(df_temp[column], kde=True, color='g');
    plt.show()
    
    plt.figure()
    print("BoxPlot of " + column)
    print("----------------------------------------------------------------------------")
    ax = sns.boxplot(x=df_temp[column])
    plt.show()
    LL, UL = remove_outlier(df_temp[column])
    print('Number of Outliers :',(df_temp[df_temp[column] > UL].shape[0] +df_temp[df_temp[column] < LL].shape[0]))
```

```python
def univariateAnalysis_category(data,cat_column):
    df_temp=data
    print("Details of " + cat_column)
    print("----------------------------------------------------------------")
    print(df_temp[cat_column].value_counts())
    plt.figure(figsize=(15,8))
    df_temp[cat_column].value_counts().plot.bar(title="Frequency Distribution of " + cat_column)
    plt.show()
    print("       ")
```

```python
def Bivariate_num_analysis(data,target,col2):
    df_temp,a,b=data,target,col2
    print('--------------------------------------------')
    print('Bivariate Analysis of {} and {}'.format(b,a))
    print('--------------------------------------------')
    print('Box Plot of {} and {}'.format(b,a))
    plt.figure(figsize=(15,8))
    sns.boxplot(data=df_temp,x=a,y=b, color='g');
    plt.show()
```

```python
def Bivariate_cat_analysis(data,target,col2):
    df_temp,a,b=data,target,col2
    print('--------------------------------------------')
    print('Bivariate Analysis of {} and {}'.format(b,a))
    print('--------------------------------------------')
    print('Count Plot between {} and {}'.format(b,a))
    sns.countplot(data=df_temp,hue=b,x=a, color='g');
    plt.show()
```

# Customer Segmentation using RFM analysis (make 4 segments)

## RFM analysis numerically ranks a customer in each of these three categories, generally on a scale of 1 to 4(for this project) (the higher the number, the better the result). The "best" customer would receive a top score in every category.
##   We Categorise the values based on quantile i.e forming bins between 0,0.25,0.50,0.75,1.0 quantiles and name them as 1,2,3,4(Except for recency where we name them as 4,3,2,1)
##  In Recency the priority is reversed as lower the value of recency higher the priority
## Then values in 3 categories are merged   444,443,442,441,434,....,112,111
## With 444 being the highest rated customer
## And 111 being the lowest rated customer
## Priority is given to the customer wit the highest rating


```python
df1=df.copy()
df1['CUSTOMER_FULL_NAME']=df1['CONTACTFIRSTNAME']+' '+df1['CONTACTLASTNAME']
print(df1['CUSTOMER_FULL_NAME'].nunique())
# df1['Current_date']=pd.to_datetime('today')
df1['Number_of_days'] = ((df1.ORDERDATE.max() - df1.ORDERDATE)/np.timedelta64(1, 'D'))
df1['Number_of_days'] = df1['Number_of_days'].astype(int)
# print(df1)
df1.head()
```

```python
t1=df1[['CUSTOMERNAME','CUSTOMER_FULL_NAME','Number_of_days']].groupby(by='CUSTOMERNAME').min()
t2=df1[['CUSTOMERNAME','SALES']].groupby(by='CUSTOMERNAME').sum()
t3=df1.groupby(by='CUSTOMERNAME')['CUSTOMERNAME'].count()
df2= pd. merge(t1,t3 , left_index=True, right_index=True)
df3= pd. merge(df2,t2 , left_index=True, right_index=True)
df3. rename( columns={'SALES':'Monetary','Number_of_days':'Recency','CUSTOMERNAME':'Frequency' } ,inplace=True) 
bin_lab=['1','2','3','4']
df3['Recency_bin']=pd.qcut(df3['Recency'],q=[0,0.25,0.50,0.75,1],labels=bin_lab[::-1])
df3['Monetary_bin']=pd.qcut(df3['Monetary'],q=[0,0.25,0.50,0.75,1],labels=bin_lab)
df3['Frequency_bin']=pd.qcut(df3['Frequency'],q=[0,0.25,0.50,0.75,1],labels=bin_lab)
df3.info()
```

```python
df3['RFM']=df3['Recency_bin'].astype(str)+df3['Frequency_bin'].astype(str)+df3['Monetary_bin'].astype(str)
df3.info()
```
